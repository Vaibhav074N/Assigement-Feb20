Q1.Explain GET and POST methods.

Ans:

The GET and POST methods are HTTP request methods used to send data from a client (such as a web browser) to a web server. Both methods are commonly used in web applications to interact with the server and retrieve or send data.

GET Method:
The GET method is used to request a resource from the server, typically to retrieve data. The data is sent as a query string appended to the URL in the form of key-value pairs. The query string is visible in the URL, making it easy to bookmark, share, or send to others. The GET method is a safe and idempotent method, which means that it does not change the state of the server and can be repeated multiple times without side effects. It is used for retrieving data that does not require any modification on the server side.

POST Method:
The POST method is used to submit data to the server to be processed or stored. The data is sent in the request body and is not visible in the URL. This method is typically used when sending sensitive or large amounts of data, such as when submitting a form or uploading a file. The POST method is not idempotent, which means that sending the same request multiple times can have different effects on the server. It is used for actions that change the state of the server, such as creating, updating, or deleting data.

Overall, the GET method is used for retrieving data, while the POST method is used for submitting data to the server for processing or storage. It's important to choose the right method based on the type of data being sent and the intended action on the server.




Q2.Why is request used in Flask?

Ans:

In Flask, the request object is used to access incoming request data. It provides a way to access the HTTP request data that is sent from the client to the server.

The request object allows you to access different attributes of the request, such as the headers, cookies, query parameters, form data, files, and more. You can use these attributes to extract the data sent by the client and use it to generate a response.

Q3.Why is redirect() used in Flask?

Ans:

In Flask, the redirect() function is used to redirect the user to a different URL. It is commonly used to redirect the user after a form submission or to redirect them to a different page based on some condition.

The redirect() function works by sending a special HTTP response to the client, called a redirect response. This response instructs the client to navigate to a different URL instead of rendering the current page

Q4. What are templates in Flask? Why is the render_template() function used?

Ans:

In Flask, templates are used to generate dynamic HTML pages that can be returned to the client as a response to a request. A template is a file that contains placeholders for dynamic content and can be populated with data from the server before it is rendered as HTML.

The render_template() function is used to render a template file and return it as a response to the client. It takes the name of the template file as an argument, along with any data that should be passed to the template. The data is passed as keyword arguments, where the name of the argument corresponds to the name of the placeholder in the template
